# API requests with urllib
En este curso, utilizará la API para una aplicación de catálogo de música. Esta API tiene múltiples características. Comenzará con la API de Letras, que le permite recuperar una cita de la Letra del día.

Antes de que pueda realizar su primera solicitud de API, necesitará saber dónde se puede acceder a la API. Esta ubicación también se conoce como URL, abreviatura de Localizador Uniforme de Recursos. La URL indicará a Python a dónde enviar la solicitud de la API. La URL de la API de Letras es la siguiente: ```http://localhost:3000/lyrics/```.

Hagamos una primera petición a la API de Lyrics utilizando el módulo ```urllib``` de Python.

## Instrucciones
* Utiliza la función ```read``` en el objeto response para leer los datos de respuesta del objeto response.
* Utiliza la función ```decode``` para decodificar los datos de respuesta en una cadena con la codificación correcta.

````python
from urllib.request import urlopen

with urlopen('http://localhost:3000/lyrics/') as response:
  
  # Use the correct function to read the response data from the response object
  data = response.read()
  encoding = response.headers.get_content_charset()

  # Decode the response data so you can print it as a string later
  string = data.decode(encoding)
  
  print(string)
````

# Uso del paquete requests
Utilizar urllib para integrar APIs puede resultar en un código verborreico y complejo, ya que necesitas ocuparte de un montón de cosas adicionales como codificar y decodificar respuestas.

Como alternativa a urllib, el paquete requests de Python ofrece una forma más sencilla de integrar APIs. Muchas de las funcionalidades están disponibles desde el principio con requests, lo que hace que tu código sea mucho más fácil de escribir y leer. Intentemos el mismo ejercicio de nuevo pero ahora con el paquete requests.

Recuerda, como en el ejercicio anterior, que la URL de la API de Lyrics es http://localhost:3000/lyrics.

## Instrucciones
* Importa el paquete ```requests```.
* Pase la URL ```http://localhost:3000/lyrics``` al método requests.get.
* Imprime el texto de la respuesta.

````python
# Import the requests package
import requests

# Pass the API URL to the get function
response = requests.get("http://localhost:3000/lyrics")

# Print out the text attribute of the response object
print(response.text)
````
